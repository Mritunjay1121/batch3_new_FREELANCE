### Task G: 10 points

Understand this,
- antecedent support: If X is called antecendent, 'antecedent support' computes the proportion of transactions that contain the antecedent X.
- consequent support: If Y is called consequent, 'consequent support' computes the proportion of transactions that contain the antecedent Y.
- support: 'support' computes the proportion of transactions that contain the antecedent X and Y.
- confidence: Probability of buying Y when X is bought.
- lift: Represents how many times the probability of getting Y increases when X is received.

Define a function which takes the (sorted values from the association rules,item) and returns top recommendation list for that item.

Tip: the recommendated item are Consequents by matching the product antecedents as item to search for.

```python


def get_top_recommendations(sorted_rules, item, top_n=5):
    # Filter association rules where the item is in antecedents
    filtered_rules = sorted_rules[sorted_rules['antecedents'].apply(lambda x: item in x)]

    # Sort the filtered rules by lift in descending order
    sorted_filtered_rules = filtered_rules.sort_values(by='lift', ascending=False)

    recommended_items = set()  # Set to store recommended items and avoid duplicates
    top_recommendations = []  # List to store top recommendations

    for consequents in sorted_filtered_rules['consequents']:
        for item_set in consequents:
            if item_set not in recommended_items:
                recommended_items.add(item_set)
                top_recommendations.append(item_set)
                if len(top_recommendations) == top_n:
                    return top_recommendations  # Return when top_n recommendations are found

    return top_recommendations

# Assuming you have already sorted_rules_by_lift from the previous code
item_to_search = 'Cake'
top_recommendations = get_top_recommendations(sorted_rules_by_lift, item_to_search)

print(f"Top 5 unique recommendations for '{item_to_search}':")
print(top_recommendations)
```

### Task M: 20 points

Submit a document showing your solution for below problem.

For the data on apriori image in the folder, apply apriori algorithm by

- Computing the support for each individual item
- Use 7 as the minimum support.
- Generate Association Rules and compute confidence 
- use min confidence as 85%.

```python
import pandas as pd
from collections import Counter

# Define the dataset
data = [
    ['beer', 'wine', 'cheese'],
    ['beer', 'potato chips'],
    ['eggs', 'flour', 'butter', 'cheese'],
    ['eggs', 'flour', 'butter', 'beer', 'potato chips'],
    ['wine', 'cheese'],
    ['potato chips'],
    ['eggs', 'flour', 'butter', 'wine', 'cheese'],
    ['eggs', 'flour', 'butter', 'beer', 'potato chips'],
    ['wine', 'beer'],
    ['beer', 'potato chips'],
    ['butter', 'eggs'],
    ['beer', 'potato chips'],
    ['flour', 'eggs'],
    ['beer', 'potato chips'],
    ['eggs', 'flour', 'butter', 'wine', 'cheese'],
    ['butter', 'wine', 'potato chips', 'cheese'],
    ['wine', 'cheese'],
    ['beer', 'potato chips'],
    ['wine', 'cheese'],
    ['beer', 'potato chips']
]

# Flatten the list of lists to get individual items
all_items = [item for sublist in data for item in sublist]

# Count the occurrences of each item
item_counts = Counter(all_items)

# Calculate support for each item
total_transactions = len(data)
item_support = {item: count / total_transactions for item, count in item_counts.items()}

# Create a DataFrame for the support values
support_df = pd.DataFrame.from_dict(item_support, orient='index', columns=['Support'])

# Sort the DataFrame by support in descending order
support_df = support_df.sort_values(by='Support', ascending=False)

print("Support for each individual item:")
print(support_df)
```

```python
# Convert the dataset to a transaction format
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm with minimum support of 7
frequent_itemsets = apriori(df, min_support=7/len(df), use_colnames=True)

# Generate Association Rules with minimum confidence of 85%
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.85)

# Print the Association Rules with confidence >= 85%
print("Association Rules with Confidence >= 85%:")
print(rules)
```
